```
df=pd.read_csv("spam.csv")
df.head()
```

```
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis='columns',inplace=True)
```

```
df.head()
```

```
df.groupby('v1').describe()
```

```
df['Spam']=df['v1'].apply(lambda x:1 if x== 'spam' else 0)
```

```
df.head()
```

```
from sklearn.feature_extraction.text import CountVectorizer
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.v2,df.Spam,test_size=0.25)
```

```
v=CountVectorizer()
X_train_count=v.fit_transform(X_train.values) # need to vectorize the strings to a binary sequence and use that for the model training
type(X_train_count)
```

```
type(X_train_count.toarray())
```

```
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count.toarray(),y_train)
```

```
email = ['Congratulations. You have won 10000 USD']
email_count=v.transform(email)
model.predict(email_count)
```
